```python
#create the weight function

    

def sigmoid(X):
    return 1/(1+np.exp(-X))    

def optimize(X,target,W,b,epochs,learning_rate):
    costs=[]
    for i in range(epochs):
        for j in range(epochs):
            m=X.shape[1]
            z=np.dot(W.T,X)+b   
            a=sigmoid(z)
            cost= -(1/m)*np.sum(target*np.log(a)+(1-target)*np.log(1-a))
            dw=(1/m)*np.dot(X,(target-a.T))
            db=(1/m)*np.sum(target-a)
   # losstrack.append(np.squeeze())
        W=W-learning_rate*dw
        b=b-learning_rate*db
        costs.append(cost)
        print("cost after %i iteration :%f" %(i,cost))
    parameter={"w":W,"b":b}
    grads={"dw":dw,"db":db}
    plt.plot(costs)
    plt.title("cost vs epochs")
    plt.xlabel("number of epochs")
    plt.ylabel("cost")
    return parameter,grads,costs
        
        #plt.plot(losstrack)
def classification(X,target,epochs,learning_rate):
   
    learning_rate=0.01 
    b = 0
    W=np.zeros((X.shape[0],1))
    parameters,grads,costs=optimize(X,target,W,b,epochs,learning_rate)
    W=parameters["w"]
    b=parameters["b"]
    d={"w":W,"b":b,"costs":costs}
    return d  

def predict(X,w,b):
    m=X.shape[1]
    print(w)
    target_value=np.zeros((1,m))
    w=w.reshape(X.shape[0],1)
    P=sigmoid(np.dot(w.T,X)+b)
    for i in range (P.shape[1]):
        if(P[0,i]<0.5):
            target_value[0,i]=0
        else:
            target_value[0,i]=1
    
    return target_value
```
